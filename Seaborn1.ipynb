```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

**Relational Plot**

```python
#scatter plot -> axes level function
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue=tips['sex'],style=tips['time'],size=tips['size'])
```

```python
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',hue=tips['sex'],style=tips['time'],size=tips['size']) # figure level function ->square shape
```

```python
sns.relplot(tips,x='total_bill',y='tip',kind='line')
```

```python
gap=px.data.gapminder()
temp_gap=gap[gap['country']=='India']
temp_gap
```

```python
#axes level function
sns.lineplot(temp_gap,x='year',y='lifeExp')
```

```python
#using relplot -> figure level function
sns.relplot(temp_gap,x='year',y='lifeExp',kind='line')
```

```python
temp_gap1=gap[gap['country'].isin(['Brazil','India','Germany'])]
temp_gap1
```

```python
sns.relplot(temp_gap1,kind='line',x='year',y='lifeExp',hue='country',style='continent',size='continent') #here size means thickness or thinness
```

```python
#facet plot => work only with figure level function
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',col='sex')
```

```python
#facet plot
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',row='sex')
```

```python
#facet plot
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',row='sex',col='smoker')
```

```python
#facet plot
sns.relplot(tips,x='total_bill',y='tip',kind='line',row='sex',col='smoker')
```

**Distribution Plot**

figure level function -> distplot

axes level function -> histplot , kdeplot, rugplot

```python
tips
```

```python
sns.histplot(tips,x='total_bill')
```

```python
sns.displot(tips,x='total_bill',kind='hist')
```

```python
sns.displot(tips,x='total_bill',kind='hist',bins=20)
```

```python
#it's also possible to visualize the distribution of a categorical variable using the logic of a histogram.
#discrete bins are automatically set for categorical data
sns.displot(tips,x='day',kind='hist')
```

```python
sns.displot(tips,x='tip',kind='hist',hue='sex')
```

```python
sns.displot(tips,x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
sns.displot(titanic,x='age',kind='hist',element='step',hue='sex')
```

```python
#faceting using col and row -> works only with figure level function
sns.displot(titanic,x='age',kind='hist',col='sex',element='step')
```

```python
#kdeplot -> kernel density estimation
#rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimation
sns.kdeplot(tips,x='total_bill')
```

```python
sns.displot(tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

**Rugplot**

this function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.


**plot marginal distributions by drawing ticks along the x and y axes.**

```python
sns.kdeplot(tips,x='total_bill')
sns.rugplot(tips,x='total_bill')
```

```python
#bivariate histogram
sns.histplot(tips,x='total_bill',y='tip')
```

```python
#bivariate kdeplot
sns.kdeplot(tips,x='total_bill',y='tip')
```

**Matrix plots**

```python
#heatmap
#plot rectangular data as a color-encoded matrix
gap
```

```python
plt.figure(figsize=(15,10))
temp=gap.pivot_table(index='country',columns='year',values='lifeExp')
sns.heatmap(temp) # axes level function
```

```python
temp1=gap[gap['continent']=='Europe'].pivot_table(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(10,20))
sns.heatmap(temp1,annot=True,linewidth=0.5,cmap='viridis')
```

```python
#clustermap
#plot a matrix dataset as a hierarchially -clustered heatmap.
#this function requires scipy to be available.
#performs clustering internally on data.
iris=px.data.iris()
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
