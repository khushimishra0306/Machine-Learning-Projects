**Boxplot**

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.


download.png

```python
sns.boxplot(tips,x='sex',y='total_bill')
```

```python
sns.boxplot(tips,x='sex',y='total_bill',hue='day')
```

```python
sns.catplot(tips,x='day',y='total_bill',hue='sex',kind='box')
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using **error bars**

```python
import numpy as np
#barplot->a categorical column with a numerical column  can calculate an aggregate value
sns.barplot(tips,x='sex',y='total_bill',errorbar=None,hue='smoker',estimator=np.min)
```

```python
sns.catplot(tips,x='sex',y='total_bill',errorbar=None,hue='smoker',estimator=np.min,kind='bar')
```

```python
#point plot
sns.pointplot(tips,x='sex',y='total_bill',hue='smoker',estimator=np.mean,errorbar=None)
```

COUNTPLOT


A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable.

```python
sns.countplot(tips,x='sex',hue='day')
```

```python
#faceting using catplot
sns.catplot(tips,x='sex',y='total_bill',col='smoker',row='time',kind='box')
```

lmplot


In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

**Multiplots**

1. Facetgrid
2. pairgrid
3.jointgrid

```python
#figure level function -> relplot,displot,catplot,lmplot
sns.catplot(tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

The TypeError in the code is because the sns.FacetGrid function does not accept x and y as direct arguments. These should be passed to a plotting function when you map it to the grid, not when you initialize the grid itself.

```python
#FacetGrid
g=sns.FacetGrid(tips,col='day',row='time')
g.map(sns.violinplot,'sex','total_bill')
```

```python
#FacetGrid
g=sns.FacetGrid(tips,col='day',row='time',hue='smoker')
g.map(sns.boxplot,'sex','total_bill')
g.add_legend()
```

```python
#plotting pairwise relationship (pairgrid vs pairplot)
#pairplot -> detects numerical coulmns automatically and (when same columns -> plot histogram)
iris
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species') #same columns here -> kde plot
```

```python
g=sns.PairGrid(iris,hue='species')
g.map(sns.scatterplot)
```

```python
# to have diff plots for diagonal and non-diagonal columns
g=sns.PairGrid(iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g=sns.PairGrid(iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# to have diff plots for diagonal and non-diagonal columns
g=sns.PairGrid(iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.histplot)
```

```python
# to have diff plots for diagonal and non-diagonal columns
g=sns.PairGrid(iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
g=sns.PairGrid(iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
