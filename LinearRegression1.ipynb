```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/placement (1).csv")
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel("cgpa")
plt.ylabel('package')
plt.title("placement analysis")
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,1:]
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
obj=LinearRegression()
obj.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
obj.predict(x_test.iloc[0].values.reshape(1,1))
```

```python
obj.predict(x_test.iloc[1].values.reshape(1,1))
```

```python
obj.predict(x_test.iloc[2].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_train,obj.predict(x_train),color='red')
plt.xlabel("cgpa")
plt.ylabel('package')
plt.title("placement analysis")
```

```python
m=obj.coef_
m
```
