```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
#Let's apply OLS
from sklearn.linear_model import LinearRegression
```

```python
obj=LinearRegression()
```

```python
obj.fit(X,y)
```

```python
obj.coef_
```

```python
obj.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,obj.predict(X),color='red')
```

```python
#Let's apply Gradient Descent assuming slope is consatnt m=78.35
#and let's assume the starting value of b =0.
y_pred=((78.35 * X )+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,obj.predict(X),color='red',label='ols')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0
loss_slope=-2* np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
#let's take learning rate=0.1
lr=0.1
step_size=loss_slope*lr
step_size
```

```python
#calculating the new intercept
b=b-step_size
b
```

```python
y_pred1=((78.35*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,obj.predict(X),color='red',label='ols')
plt.plot(X,y_pred1,color='blue',label='b=0')
plt.plot(X,y_pred,color='green',label='b={}'.format(b))
plt.legend()
plt.show()
```

```python
#iteration 2
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=lr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=((78.35*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,obj.predict(X),color='red',label='ols')
plt.plot(X,y_pred1,color='blue',label='b=0')
plt.plot(X,y_pred2,color='black',label='b={}'.format(b))
plt.legend()
plt.show()
```

```python
#iteration 3
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=lr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred3=((78.35*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,obj.predict(X),color='red',label='ols')
plt.plot(X,y_pred1,color='blue',label='b=0')
plt.plot(X,y_pred3,color='black',label='b={}'.format(b))
plt.legend()
plt.show()
```

```python
b=-100
m=78.35
lr=0.1
epochs=10
for i in range(epochs):
  loss_slope=-2* np.sum(m*X.ravel()-b)
  b=b-(lr*loss_slope)
  y_pred=m*X+b
  plt.plot(X,y_pred)
plt.scatter(X,y)
```
