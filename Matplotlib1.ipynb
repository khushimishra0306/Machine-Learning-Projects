```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
plt.style.use('default')
```

```python
df=pd.read_csv("/content/sharma-kohli.csv")
df
```

```python
plt.plot(df['index'],df['V Kohli'],color="green",linestyle="solid",linewidth=3,marker='D')
plt.plot(df['index'],df['RG Sharma'],color="blue",linestyle="dashdot",linewidth=2,marker='o',markersize=5)
plt.title("Rohit Sharma VS Virat Kohli career comparison")
plt.xlabel("season of IPL")
plt.ylabel("Runs ")
```

```python
plt.plot(df['index'],df['V Kohli'],color="green",linestyle="solid",linewidth=3,marker='D',label="virat")
plt.plot(df['index'],df['RG Sharma'],color="blue",linestyle="dashdot",linewidth=2,marker='o',markersize=5,label="rohit")
plt.title("Rohit Sharma VS Virat Kohli career comparison")
plt.xlabel("season of IPL")
plt.ylabel("Runs ")
plt.legend(loc="best")
plt.grid()
plt.show() # the main func jiski vajah se graph plot hote h.
#upper right
#upper left
#lower left
#lower right
#right
#center left
#center right
#lower center
#upper center
```

```python
#multiple bar charts
df2=pd.read_csv("/content/batsman_season_record.csv")
df2
```

```python
plt.bar(df2['batsman'],df2['2015'])
plt.bar(df2['batsman'],df2['2016'])
plt.bar(df2['batsman'],df2['2017'])
```

```python
#to solve above problem

plt.bar(np.arange(df2.shape[0])-0.2,df2['2015'],width=0.2,color="pink")
plt.bar(np.arange(df2.shape[0]),df2['2016'],width=0.2,color="red")
plt.bar(np.arange(df2.shape[0])+0.2,df2['2017'],width=0.2,color="black")
plt.xticks(np.arange(df2.shape[0]),df2['batsman'])
plt.show()
```
