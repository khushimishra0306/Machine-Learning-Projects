**Making Our Own Regression Class**

```python
class MeraLinearRegression:
  def __init(self):
    self.coef=None
    self.intercept=None
  def fit(self,X_train,Y_train):
    X_train=np.insert(X_train,0,1,axis=1) # kis dataset pe krne h changes, index,kya changes,column level/row level
    betas=np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(Y_train)
    print(betas)
    self.intercept=betas[0]
    self.coef=betas[1:]
  def predict(self,X_test):
    y_pred=np.dot(X_test,self.coef)+self.intercept
    print(y_pred.shape)
    return y_pred




```

```python
myObj=MeraLinearRegression()
myObj.fit(X_train,y_train)
```

```python
Y_pred=myObj.predict(X_test)
print(Y_pred)
```

```python
R21=r2_score(y_test,Y_pred)
print(R21)
```
